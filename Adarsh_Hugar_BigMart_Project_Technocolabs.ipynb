### Feature Engineering

```python
# Converting catergorical to numeric
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
encoder = LabelEncoder()
```

```python
cpy_data.head(2)
```

```python
cpy_data['E_Item_Fat_Content'] =encoder.fit_transform(cpy_data['Item_Fat_Content'])
```

```python
cpy_data['E_Item_Type'] =encoder.fit_transform(cpy_data['Item_Type'])
cpy_data['E_Outlet_Identifier'] =encoder.fit_transform(cpy_data['Outlet_Identifier'])
cpy_data['E_Outlet_Size'] =encoder.fit_transform(cpy_data['Outlet_Size'])
cpy_data['E_Outlet_Location_Type'] =encoder.fit_transform(cpy_data['Outlet_Location_Type'])
cpy_data['E_Outlet_Type'] =encoder.fit_transform(cpy_data['Outlet_Type'])
```

```python
cpy_data
```

```python
processed_data = cpy_data.drop(['Item_Identifier','Item_Type','Item_Fat_Content','Outlet_Identifier','Outlet_Size','Outlet_Location_Type','Outlet_Type'],axis=1)
```

```python
processed_data
```

```python
processed_data
```

```python
processed_data.isna().sum()
```

```python
processed_data.fillna(processed_data.mean(), inplace=True)
```

```python
processed_data.isna().sum()
```

```python
###  Splitiing the data into input and output features
```

```python
X_unscaled = processed_data.drop(['Item_Outlet_Sales'],axis=1)
y_unscaled = processed_data['Item_Outlet_Sales']
```

```python
X_unscaled.head(2)
```

```python
y_unscaled.head(2)
```

```python
# Splitiing the data into training and testing

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_unscaled, y_unscaled, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
```

```python
model_lr = LinearRegression()
model_lr.fit(X_unscaled, y_unscaled)
```

```python
# Making predictions

unscaled_linear_y_pred = model_lr.predict(X_test)
```

```python
unscaled_linear_y_pred
```

### Model Evaluation

```python
from sklearn.metrics import mean_squared_error,r2_score
```

```python
mse_unscaled_linear = mean_squared_error(y_test, unscaled_linear_y_pred)
mse_unscaled_linear
```

```python
r2_unscaled_linear = r2_score(y_test, unscaled_linear_y_pred)
```

```python
r2_unscaled_linear
```

## `Insights` 


- As we can observe the linear regression is giving accuracy of 52.13% which is very less.
- This can be due to the data which is not scaled down.High range in numbers can efffect the model.

- Lets build the model by scaling down the data and check the accuracy

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_scaled = scaler.fit_transform(X_unscaled)
X_scaled
```

```python
y_unscaled
```

### Splitting train and test data on scaled data

```python
X_train_scaled, X_test_scaled, y_train, y_test = train_test_split(X_scaled, y_unscaled, test_size=0.2, random_state=42)
```

#### Building model on scaled data

```python
model_lr.fit(X_train_scaled,y_train)
```

```python
scaled_linear_y_pred = model_lr.predict(X_test_scaled)
scaled_linear_y_pred
```

```python
mse_scaled_linear = mean_squared_error(y_test, scaled_linear_y_pred)
mse_scaled_linear
```

```python
r2_scaled_linear = r2_score(y_test, scaled_linear_y_pred)
r2_scaled_linear
```

`Insights`

- Even after scaling down the data accuracy is not improved.

```python
### Model using Randomforest
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_rf = RandomForestRegressor(n_estimators=100, random_state=42)
model_rf.fit(X_scaled, y_unscaled)
```

```python
# Make predictions
y_pred_rf = model_rf.predict(X_test_scaled)

# Evaluate the model
mse_rf = mean_squared_error(y_test, y_pred_rf)
mse_rf
```

```python
# Evaluate the model
mse_rf = mean_squared_error(y_test, y_pred_rf)
r2_scaled_RF = r2_score(y_test, y_pred_rf)
```

```python
print('The Mean squared error for Random Forest model is {}'.format(mse_rf))
print('The R squaared value for Random Forest model is {}'.format(r2_scaled_RF))
```

#### The accuracy of model when built with Random Forest is 93.8%

```python
model_rf_6 = RandomForestRegressor(n_estimators=100, random_state=42,max_depth=6)
model_rf_6.fit(X_scaled, y_unscaled)
```

```python
y_pred_rf_6 = model_rf_6.predict(X_test_scaled)
```

```python
mse_rf_6 = mean_squared_error(y_test, y_pred_rf_6)
r2_scaled_RF_6 = r2_score(y_test, y_pred_rf_6)
```

```python
print('The Mean squared error for Random Forest model is {}'.format(mse_rf_6))
print('The R squaared value for Random Forest model is {}'.format(r2_scaled_RF_6))
```

```python
### Pruning the tree for chceking 
```

```python
model_rf_5 = RandomForestRegressor(n_estimators=100, random_state=42,max_depth=5)
model_rf_5.fit(X_scaled, y_unscaled)

y_pred_rf_5 = model_rf_5.predict(X_test_scaled)

mse_rf_5 = mean_squared_error(y_test, y_pred_rf_5)
r2_scaled_RF_5 = r2_score(y_test, y_pred_rf_5)

print('The Mean squared error for Random Forest model is {}'.format(mse_rf_5))
print('The R squaared value for Random Forest model is {}'.format(r2_scaled_RF_5))
```

```python
## We can see model accuracy is decreasing with less depth.
```
